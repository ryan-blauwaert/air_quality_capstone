```python
import numpy as np
import pandas as pd
```

```python
air_qual_1999 = pd.read_csv('../data/conreport1999.csv')
air_qual_1999.head()
```

```python
air_qual_1999.info()
```

```python
numeric_cols = air_qual_1999.columns[2:]
numeric_cols
```

```python
air_qual_1999[numeric_cols] = air_qual_1999[numeric_cols].replace('.', np.nan)
```

```python
air_qual_1999.head()
```

```python
air_qual_1999.info()
```

```python
for col in numeric_cols:
    air_qual_1999[col] = pd.to_numeric(air_qual_1999[col])
```

```python
air_qual_1999.info()
```

```python
# rename two locations manually for consistency with inf_mor dataframe
air_qual_1999['CBSA'] = air_qual_1999['CBSA'].str.replace('Washington-Arlington-Alexandria, DC-VA-MD-WV', 'District of Columbia')
air_qual_1999['CBSA'] = air_qual_1999['CBSA'].str.replace('New York-Newark-Jersey City, NY-NJ-PA', 'New York City')
```

```python
# try to make 'CBSA' column match format of Locations in inf_mor dataframe
air_qual_1999['CBSA'] = air_qual_1999['CBSA'].str.split(',', expand=True)
```

```python
air_qual_1999['CBSA'] = air_qual_1999['CBSA'].str.split('/', expand=True)
```

```python
air_qual_1999['CBSA'] = air_qual_1999['CBSA'].str.split('-', expand=True)
```

```python
air_qual_1999.head()
```

```python
# import infant mortality dataframe 
inf_mor = pd.read_csv('../data/infant_mortality.csv')
```

```python
# list of cities included in inf_mor dataframe
inf_mor_cities = inf_mor['Location'].unique()
inf_mor_cities
```

```python
# check for cities to drop from inf_mor dataframe (no corresponding air_quality data)
cities_to_drop = [city for city in inf_mor_cities if city not in air_qual_1999['CBSA'].unique()]
cities_to_drop
```

```python
# drop these rows from inf_mor data and rewrite
inf_mor = inf_mor.loc[~(inf_mor['Location'].isin(cities_to_drop))]
inf_mor.info() #now showing correct number of entries
```

```python
inf_mor.to_csv('../data/infant_mortality.csv')
```

```python
cities = inf_mor['Location'].unique()
cities
```

```python
air_qual_1999 = air_qual_1999.loc[air_qual_1999['CBSA'].isin(cities)]
air_qual_1999
```

```python
# investigate duplicates manually
cbsa_codes_to_drop = [17420, 17980, 27340, 38860]
air_qual_1999 = air_qual_1999.loc[~(air_qual_1999['CBSA Code'].isin(cbsa_codes_to_drop))]
air_qual_1999
```

```python
air_qual_1999.info() #dataframe lengths now match
```

```python
air_qual_1999.to_csv('../data/cleaned/air_1999.csv')
```
