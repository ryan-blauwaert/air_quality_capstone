```python
asthma_nyc = pd.read_csv('../data/asthma_hospitalizations_nyc.csv')
asthma_nyc
```

```python
asthma_nyc = asthma_nyc.loc[asthma_nyc['Location'] == 'New York City']
asthma_nyc.loc[asthma_nyc['DataFormat'] == 'Rate']
```

```python
asthma_nyc = asthma_nyc.loc[(asthma_nyc['Age Group'] == '0 to 4 years') & (asthma_nyc['DataFormat'] == 'Rate')]
asthma_nyc['TimeFrame'] = asthma_nyc['TimeFrame'].astype(str)
```

```python
asthma_nyc = asthma_nyc[['TimeFrame', 'Data']].set_index('TimeFrame', drop=True).transpose()
```

```python
asthma_nyc.rename({'Data': 'Asthma Hosp. Rate'}, axis=0, inplace=True)
```

```python
nyc_data = nyc_data.append(asthma_nyc.loc['Asthma Hosp. Rate'])
```

```python
nyc_data
```

```python
nyc_data.corrwith(nyc_data.loc['Asthma Hosp. Rate'], axis=1)
```

```python
fig, ax = plt.subplots(figsize=(14,6))
ax.plot(nyc_data.columns, nyc_data.loc['CO 2nd Max 1-hr'], label='Carbon Monoxide')
ax.plot(nyc_data.columns, nyc_data.loc['SO2 Mean 1-hr'], label='Sulfur Dioxide')
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(6, 1, figsize=(12, 15), sharex=True)
[ax1, ax2, ax3, ax4, ax5, ax6] = axs.flatten()
ax1.plot(nyc_data.columns, nyc_data.loc['CO 2nd Max 1-hr'], label='Carbon Monoxide')
ax2.plot(nyc_data.columns, nyc_data.loc['SO2 Mean 1-hr'], label='Sulfur Dioxide')
ax3.plot(nyc_data.columns, nyc_data.loc['Ozone 2nd Max 1-hr'], label='Ozone')
ax4.plot(nyc_data.columns, nyc_data.loc['PM2.5 Weighted Mean 24-hr'], label='Particulate Matter')
ax5.plot(nyc_data.columns, nyc_data.loc['Infant Mortality'], label='Infant Mortality')
ax6.plot(nyc_data.columns, nyc_data.loc['Asthma Hosp. Rate'], label='Childhood Asthma Hospitalizations')
fig.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(6,1, figsize=(12,20), sharex=True)
metrics = ['CO 2nd Max 1-hr', 'SO2 Mean 1-hr', 'Ozone 2nd Max 1-hr', 'PM2.5 Weighted Mean 24-hr', 'Infant Mortality', 'Asthma Hosp. Rate']
titles = ['Carbon Monoxide','Sulfur Dioxide','Ozone','Particulate Matter','Infant Mortality','Childhood Asthma Hospitalizations']
palette = itertools.cycle(sns.color_palette())
y_labels = ['Concentration, ppm', 'Concentration, ppb', 'Concentration, ppm', 'Concentration, ug/m3', 'Deaths per 1,000 Births', 'Hosp. per 10,000 Children']
for i, ax in enumerate(axs.flatten()):
    ax.plot(nyc_data.columns, nyc_data.loc[metrics[i]], color=next(palette), marker='o', lw=3)
    ax.set_title(titles[i], size=16)
    ax.set_ylabel(y_labels[i], size=16)
ax.set_xlabel('Year', size=16)
fig.suptitle('New York City Air Pollution and Health Outcomes', x=.52, y=.99, size=20)
fig.tight_layout()
# plt.savefig('../images/nyc_air_and_health.png)
plt.show()
```
