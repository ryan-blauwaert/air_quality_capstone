```python
def add_infant_mortality_row(file_location, other_df, row_to_add, new_row_name):
    df = pd.read_csv(file_location, index_col='Unnamed: 0')
    df = df.append(other_df.loc[row_to_add])
    df = df.rename(index={row_to_add: new_row_name})
    return df
```

```python
add_infant_mortality_row('../data/cleaned/new_york.csv', inf_mor, 'New York City', 'Infant Mortality')
```
