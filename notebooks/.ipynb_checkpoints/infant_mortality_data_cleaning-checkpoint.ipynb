```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
inf_mor_1999 = pd.read_csv('../data/infant_mortality_1999_2008.txt')
inf_mor_2009 = pd.read_csv('../data/infant_mortality_2009_2018.txt')
```

```python
inf_mor_1999.head()
```

```python
inf_mor_2009.head()
```

```python
cities_1999 = inf_mor_1999['Location'].unique()
```

```python
# only interested in cities that appear in both datasets
inf_mor_2009 = inf_mor_2009.loc[inf_mor_2009['Location'].isin(cities_1999)]
inf_mor_2009.head()
```

```python
inf_mor_2009.reset_index(drop=True, inplace=True)
```

```python
inf_mor_2009.head()
```

```python
# merge two datasets into one to show 20 years
inf_mor = inf_mor_1999.merge(inf_mor_2009, how='inner', on='Location')
```

```python
inf_mor
```

```python
inf_mor.info()
```

```python
# column Data_Type_y is redundant; drop it
inf_mor.drop('Data Type_y', axis=1, inplace=True)
```

```python
# confirm drop
inf_mor.info()
```

```python
# Rename Data Type_x to Data Type
inf_mor.columns.rename({'Data Type_x': 'data_type'}, inplace=True)
```

```python
# need to replace values causing numeric columns to appear as dtype=object; first check unique values
inf_mor['1999'].unique()
```

```python
# notice that 'N.A.' appears; replace all instances with np.nan
inf_mor.replace('N.A.', np.nan, inplace=True)
inf_mor.info()
```

```python
# try to cast columns to floats
years = inf_mor.columns[2:]
for year in years:
    inf_mor[year] = pd.to_numeric(inf_mor[year])
inf_mor.info()
```

```python
# replace 'S' win np.nan
inf_mor.replace('S', np.nan, inplace=True)
```

```python
inf_mor.info()
```

```python
# try again
for year in years:
    inf_mor[year] = pd.to_numeric(inf_mor[year])
inf_mor.info()
```

```python
inf_mor.describe()
```

```python
inf_mor
```

```python
# drop '-Davidson' for index 30 for consistency
inf_mor.loc[30, 'Location'] = 'Nashville'
```

## Making air quality data cleaning pipeline

```python
air_qual_1999 = pd.read_csv('../data/conreport1999.csv')
air_qual_1999
```

```python
# manually replace some city names for consistency
air_qual_1999.loc[air_qual_1999['CBSA'].str.contains('DC')]
air_qual_1999['CBSA'] = air_qual_1999['CBSA'].str.replace('Washington-Arlington-Alexandria, DC-VA-MD-WV', 'District of Columbia')
```

```python
air_qual_1999.loc[air_qual_1999['CBSA'].str.contains('New York')]
air_qual_1999['CBSA'] = air_qual_1999['CBSA'].str.replace('New York-Newark-Jersey City, NY-NJ-PA', 'New York City')
```

```python
# Try to make 'CBSA' column match format of cities in infant mortality dataframe
air_qual_1999['CBSA'] = air_qual_1999['CBSA'].str.split(',', expand=True)
```

```python
air_qual_1999.head()
```

```python
# drop secondary cities after hyphen in CBSA column
air_qual_1999['CBSA'] = air_qual_1999['CBSA'].str.split('-', expand=True)
air_qual_1999.head(10)
```

```python
# get cities to research further
cities_not_in_air_qual = []
for city in inf_mor['Location']:
    if city not in air_qual_1999['CBSA'].unique():
        cities_not_in_air_qual.append(city)
cities_not_in_air_qual
```

```python
# All cities except Lousville encompassed in data from neighboring cities; fix Louisville:
air_qual_1999.loc[air_qual_1999['CBSA'].str.contains('Louisville')]
```

```python
air_qual_1999['CBSA'] = air_qual_1999['CBSA'].str.split('/', expand=True)
```

```python
# try again
cities_not_in_air_qual = []
for city in inf_mor['Location']:
    if city not in air_qual_1999['CBSA'].unique():
        cities_not_in_air_qual.append(city)
cities_not_in_air_qual
```

```python
# drop these cities from inf_mor since data is found elsewhere
inf_mor = inf_mor.loc[~(inf_mor['Location'].isin(cities_not_in_air_qual))]
inf_mor.info()
```

```python
# rewrite inf_mor to csv
inf_mor.to_csv('../data/infant_mortality.csv')
```
