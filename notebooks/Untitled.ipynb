```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import itertools
plt.style.use('seaborn-darkgrid')
palette = itertools.cycle(sns.color_palette())
```

```python
air = pd.read_csv('../data/cleaned/air_all_years.csv', index_col='Year')
```

```python
air
```

```python
air.index.unique()
```

```python
co_1999 = air.loc[1999, 'CO 2nd Max 1-hr']
```

```python
co_2018 = air.loc[2018, 'CO 2nd Max 1-hr']
```

```python
co_2018 = co_2018[~(np.isnan(co_2018))]
```

```python
co_2009 = air.loc[2009, 'CO 2nd Max 1-hr']
co_2009 = co_2009[~(np.isnan(co_2009))]
```

```python
co_2018
```

```python
fig, axs = plt.subplots(1,3, figsize=(12,8), sharey=True)
ax1, ax2, ax3 = axs.flatten()
ax1.boxplot(co_1999)
ax1.set_xlabel('1999', size=20)
ax2.boxplot(co_2009)
ax2.set_xlabel('2009', size=20)
ax3.boxplot(co_2018)
ax3.set_xlabel('2018', size=20)
ax1.set_ylabel('Concentration, ppm', size=20)
plt.suptitle('National Carbon Monoxide 1999, 2009, 2018', size=32)
fig.tight_layout()
plt.savefig('../images/eda_boxplot_example')
plt.show()
```
